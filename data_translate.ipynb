```
import pandas as pd
from deep_translator import GoogleTranslator
```

```
dfs = {}
```

```
translator = GoogleTranslator(source='cs', target='en')
```

```
dfs['dnn_cis'] = pd.read_csv('/Users/mihailborisov/Desktop/portfolio/eco_expenses/czech/dnn_cis.csv', encoding='utf-8')
dfs['ozp_cis'] = pd.read_csv('/Users/mihailborisov/Desktop/portfolio/eco_expenses/czech/ozp_cis.csv', encoding='utf-8')
dfs['sector_cis'] = pd.read_csv('/Users/mihailborisov/Desktop/portfolio/eco_expenses/czech/sector_cis.csv', encoding='utf-8')
```

```
def build_translation_map(series):

    unique_vals = series.dropna().unique()

    translations = {}

    for val in unique_vals:
        if isinstance(val, str):
            val_strip = val.strip()

            if val_strip.isdigit():
                translations[val] = val
            else:
                try:
                    translations[val] = translator.translate(val_strip)
                except:
                    translations[val] = val
        else:
            translations[val] = val

    return translations
```

```
for key in dfs.keys():
    df = dfs[key]

    for col in df.select_dtypes(include="object").columns:
        translation_map = build_translation_map(df[col])
        df[col] = df[col].map(translation_map)

    df.to_csv(
        f'/Users/mihailborisov/Desktop/portfolio/eco_expenses/eng/{key}_eng.csv',
        index=False,
        encoding='utf-8'
    )
```
